```python
def searchUnreachible(file):
    mas = []
    with open(file, 'r', encoding='utf-8') as f:  
        lines = f.readlines()
        for line in lines: 
            mas.append(list(map(int, line.split())))
    print(mas)
    
    result = []
    for i in range(len(mas)):
        for j in range(len(mas[i])):
            if mas[i][j] == 1:
                break
            elif j == len(mas[i]) - 1:
                result.append(i + 1)
    return result
```
